# Análisis Exploratorio de Datos (EDA)
En esta sección visualizamos las tendencias principales de los apartamentos recolectados.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')
```

```python
# 1. Distribución de Precios de Canon
plt.figure(figsize=(10, 6))
sns.histplot(df['precio canon'], kde=True, color='teal')
plt.title('Distribución del Precio Canon')
plt.xlabel('Precio (COP)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# 2. Precio promedio por Zona
plt.figure(figsize=(12, 6))
df.groupby('zona')['precio canon'].mean().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Precio Promedio por Zona')
plt.ylabel('Precio Promedio (COP)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Relación entre Metros Cuadrados y Precio
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='metros_cuadrados', y='precio canon', hue='habitaciones', palette='viridis')
plt.title('Relación Área vs Precio')
plt.xlabel('Área (m²)')
plt.ylabel('Precio (COP)')
plt.show()
```

```python
# 4. Correlación entre variables numéricas
plt.figure(figsize=(8, 6))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor de Correlaciones')
plt.show()
```
